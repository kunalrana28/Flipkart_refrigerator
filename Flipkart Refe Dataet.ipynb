```python
from selenium import webdriver as wb 
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
import time
import csv
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
webD=wb.Chrome(executable_path="D:\Chrome Driver\chromedriver.exe")
```

```python
Box = []
Type = []
RType = []
DType = []
CType = []
Capacity = []
NDoors = []
SRating = []
Coolpad = []
TGlass = []
BStabilizer = []
```

```python
Name = []
Price = []
Rating = []
for i in range(1,41):
    webD.get('https://www.flipkart.com/search?q=refrigerator&sid=j9e%2Cabm%2Chzg&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=refrigerator%7CRefrigerators&requestId=9e85b4e7-5d02-42e1-bc47-17372832f7b4&as-searchtext=refe&page='+str(i)+'')
    for j in range(2,26):
        try:
            name = webD.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div['+str(j)+']/div/div/div/a/div[2]/div[1]/div[1]')
            Name.append(name.text)
        except NoSuchElementException:
            Name.append(None)
        
        try:
            price = webD.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div['+str(j)+']/div/div/div/a/div[2]/div[2]/div[1]/div/div[1]')
            Price.append(price.text)
        except NoSuchElementException:
            Price.append(None)
        try:
            rating = webD.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div['+str(j)+']/div/div/div/a/div[2]/div[1]/div[2]/span[1]/div')
            Rating.append(rating.text)
        except NoSuchElementException:
            Rating.append(None)
        link=webD.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div['+str(j)+']/div/div/div/a').get_attribute('href')
        webD.find_element_by_xpath('/html/body/div[1]/div/div[3]/div[1]/div[2]/div['+str(j)+']/div/div/div/a/div[2]/div[1]/div[1]').click()
        webD.switch_to.window(webD.window_handles[1])
        time.sleep(1)
        req=requests.get(link).text
        soup = BeautifulSoup(req,'html.parser')
        td=[]
        th=[]

        commonclass = soup.find_all('table', class_="_14cfVK")
        l1=commonclass[0].find_all('tr')
        for i in range(0,len(l1)):
            t=l1[i].find_all('td')
            u=t[0].text
            v=t[1].text

            td.append(u)
            th.append(v)
        dictionary = dict(zip(td,th))
        
        
        key1='In The Box'
        key2='Type'
        key3='Refrigerator Type'
        key4='Defrosting Type'
        key5='Compressor Type'
        key6='Capacity'
        key7='Number of Doors'
        key8='Star Rating'
        key9='Coolpad'
        key10 = 'Toughened Glass'
        key11 = 'Built-in Stabilizer'
        
        if key1 in dictionary.keys():
            Box.append(dictionary[key1])
        else:
            Box.append(None)

        if key2 in dictionary.keys():
            Type.append(dictionary[key2])
        else:
            Type.append(None)


        if key3 in dictionary.keys():
            RType.append(dictionary[key3])
        else:
            RType.append(None)

        if key4 in dictionary.keys():
            DType.append(dictionary[key4])
        else:
            DType.append(None)    

        if key5 in dictionary.keys():
            CType.append(dictionary[key5])
        else:
            CType.append(None) 
        if key6 in dictionary.keys():
            Capacity.append(dictionary[key6])
        else:
            Capacity.append(None)


        if key7 in dictionary.keys():
            NDoors.append(dictionary[key7])
        else:
            NDoors.append(None)

        if key8 in dictionary.keys():
            SRating.append(dictionary[key8])
        else:
            SRating.append(None) 
        
        if key9 in dictionary.keys():
            Coolpad.append(dictionary[key9])
        else:
            Coolpad.append(None)
        if key10 in dictionary.keys():
            TGlass.append(dictionary[key10])
        else:
            TGlass.append(None)
        if key11 in dictionary.keys():
            BStabilizer.append(dictionary[key11])
        else:
            BStabilizer.append(None)
        webD.close()
        webD.switch_to.window(webD.window_handles[0])
```

```python
df = {
    'In The Box' : Box,                                     
    'Type' :Type,
    'Refrigerator Type' : RType,
    'Defrosting Type' : DType,
    'Compressor Type': CType,
    'Capacity' : Capacity,
    'Number of Doors' : NDoors,
    'Star Rating' : SRating,
    'Coolpad' : Coolpad,
    'Toughened Glass' : TGlass,
    'Built-in Stabilizer' : BStabilizer,   
}
FlipKart_refe_dataset=pd.DataFrame(data=df)
```
